```python
#Selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


#Time
from time import sleep

import pandas as pd
```

```python
dfDonors = pd.read_csv("past_donors/past_donors_clean.csv")
```

```python
dfDonors.head()
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://linkedin.com')
```

```python
def setupLinkedIn(email, password,sleep_time):
    #Setup   
    selEmail = driver.find_element(By.ID, 'session_key')
    selPW = driver.find_element(By.ID, 'session_password')
    selEmail.send_keys(email)
    selPW.send_keys(password)
    selPW.send_keys(Keys.RETURN)

    sleep(sleep_time)
```

```python
def searchCompany(company_name, sleep_time):
    #Click on search button
    searchButton = driver.find_element(By.CSS_SELECTOR, "input[aria-label='Search']")
    searchButton.click()

    #Search

    searchButton.send_keys(company_name)
    searchButton.send_keys(Keys.RETURN)

    sleep(sleep_time)
```

```python
def getAboutText(sleep_time):
    #Click on link
    company_link = driver.find_element(By.CSS_SELECTOR, ".app-aware-link [href]").get_attribute("href")
    about_link = company_link + 'about/'
    driver.get(about_link)

    sleep(sleep_time)
    aboutText = driver.find_element(By.ID, "ember58")
    
    return aboutText
```

```python
def returnHomePage(sleep_time):
    driver.get('https://www.linkedin.com/feed/')
    sleep(sleep_time)
```

```python
def extractLinkedInData(email, password, company_names):
    temp_list = []
    setupLinkedIn(email, password, 5)
    
    for company_name in company_names:
        searchCompany(company_name, 5)
        
        temp_list.append(getAboutText(5))
        
        returnHomePage(5)
    
    return temp_list
```
