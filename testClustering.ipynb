```python
#Define string cleaning function
def cleanText(text):
    text = str(text).lower()
    
    text = re.sub(r"\s+", " ", text)  # Remove multiple spaces in content

    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  # Remove punctuation
    
    text = re.sub(r"\s+", " ", text)  # Remove multiple spaces in content
    
    return text
```
